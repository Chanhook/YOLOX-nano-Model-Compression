```python
# 문자열 파일 순서 정렬 library
# pip install natsort
import natsort

# test dataset annotation path를 입력해주세요
annotations_path = (
    "../datasets/test/VOCdevkit/VOC2007/Annotations"
)
annotations = natsort.natsorted(os.listdir(annotations_path))
annotations
```

```python
# xml dictionary 형식으로 바꿔주는 library
# pip install xmltodict
import xmltodict
import json

anno_list = []
for anno in annotations:
    with open(os.path.join(annotations_path, anno), encoding="utf-8") as xml_file:
        data_dict = xmltodict.parse(xml_file.read())
    xml_file.close()
    json_string = json.dumps(data_dict)
    json_data = json.loads(json_string)
    anno_list.append(json_data)
```

```python
IMAGE_EXT = [".jpg", ".jpeg", ".webp", ".bmp", ".png"]


def get_image_list(path):
    image_names = []
    for maindir, subdir, file_name_list in os.walk(path):
        for filename in file_name_list:
            apath = os.path.join(maindir, filename)
            ext = os.path.splitext(apath)[1]
            if ext in IMAGE_EXT:
                image_names.append(apath)
    return image_names
```

```python
img_path = (
    "../datasets/test/VOCdevkit/VOC2007/JPEGImages"
)

if os.path.isdir(img_path):
    image_files = get_image_list(img_path)
else:
    image_files = [img_path]
image_files.sort()
image_files
```

```python
__all__ = ["vis"]

cls_names = (
    "aeroplane",
    "bicycle",
    "bird",
    "boat",
    "bottle",
    "bus",
    "car",
    "cat",
    "chair",
    "cow",
    "diningtable",
    "dog",
    "horse",
    "motorbike",
    "person",
    "pottedplant",
    "sheep",
    "sofa",
    "train",
    "tvmonitor",
)


def drawing(obj, class_names):
    box = obj["bndbox"]
    cls_id = class_names.index(obj["name"])

    x0 = int(box["xmin"])
    y0 = int(box["ymin"])
    x1 = int(box["xmax"])
    y1 = int(box["ymax"])

    color = (_COLORS[cls_id] * 255).astype(np.uint8).tolist()
    text = "{}".format(class_names[cls_id])
    txt_color = (0, 0, 0) if np.mean(_COLORS[cls_id]) > 0.5 else (255, 255, 255)
    font = cv2.FONT_HERSHEY_SIMPLEX

    txt_size = cv2.getTextSize(text, font, 0.4, 1)[0]
    cv2.rectangle(img, (x0, y0), (x1, y1), color, 2)

    txt_bk_color = (_COLORS[cls_id] * 255 * 0.7).astype(np.uint8).tolist()
    cv2.rectangle(
        img,
        (x0, y0 + 1),
        (x0 + txt_size[0] + 1, y0 + int(1.5 * txt_size[1])),
        txt_bk_color,
        -1,
    )
    cv2.putText(img, text, (x0, y0 + txt_size[1]), font, 0.4, txt_color, thickness=1)


def vis(img, objs_info, class_names=cls_names):
    if type(objs_info) != type([]):
        drawing(objs_info, class_names)
    else:
        for obj in objs_info:
            drawing(obj, class_names)
    return img
```

```python
save_folder = "../gt_output/"

for image_file, anno in zip(image_files, anno_list):
    img = cv2.imread(image_file)

    objs_info = anno["annotation"]["object"]
    result_img = vis(img, objs_info)

    os.makedirs(save_folder, exist_ok=True)
    save_file_name = os.path.join(save_folder, os.path.basename(image_file))
    print("Saving detection result in {}".format(save_file_name))
    cv2.imwrite(save_file_name, result_img)
print("Saving ground truth images is done!")
```
